```python
import whit_phys_util as buc
import numpy as np
import matplotlib.pyplot as plt
```

```python
repo = buc.local_repository("https://github.com/cambo2018/Advanced-Dynamics")
```

```python
t = np.linspace(0,10,10000)

beta = 0.5
delta = 0
w1 = 2*np.pi

x = np.exp(-beta*t)*np.cos(w1*t - delta)
xdot = (-beta)*np.exp(-beta*t)*np.cos(w1*t - delta) - np.exp(-beta*t)*np.sin(w1*t - delta)*(w1)
```

```python
fig, ax = plt.subplots()
ax.plot(x,xdot)
ax.set_xlabel('x')
ax.set_ylabel('xdot');
```

The attractor here is a stable spiral, as at all points, the flow is towards the center of the spiral.  Energy is not conserved.  If it was, a closed loop would be formed.  This could be the phase space for a damped pendulum, where is slowly damps down to (x,xdot) = (0,0).
