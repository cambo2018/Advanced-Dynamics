```python
# Given in problem
B = .1 #changed from .2 in Ex. 5.5
T_0 = 1
w_0 = 2*pi/T_0
T = 1 #, 1.5, 2, 2.5
w = 2*pi/T
dT = .25
f_max = 1
```

```python
var('t')
f = Piecewise( (0, t < -dT/2), (0, t > +dT/2), (f_max, True))
f
```

```python
fig1 = plot(f, (t, -T/2, T/2), line_color="red", label="exact")
```

```python
var('n', integer=True, positive=True)

f0 = integrate((1/T)*f, (t, -T/2, T/2))

def fn(n):
  return integrate(2/T*cos(n*w*t)*f, (t, -T/2, T/2))

def f_approx(nmax, t):
  f_sym = f0 + Sum(fn(n)*cos(n*w*t), (n, 1, nmax))
  return f_sym.doit()
```

```python
fig2 = plot(f_approx(1, t), (t, -T/2, T/2), show = False, line_color="green", label="n = 1")
fig3 = plot(f_approx(10, t), (t, -T/2, T/2), show = False, line_color="blue", label="n = 10")
fig1.legend = True
fig1.extend(fig2)
fig1.extend(fig3)
fig1.show()
```

```python
T2 = 1.5
T3 = 2
T4 = 2.5

w2 = 2*pi/T2
w3 = 2*pi/T3
w4 = 2*pi/T4

a0 = f0/w_0**2
def A(n, w):
  return fn(n)/sqrt((w_0**2 - (n*w)**2)**2 + (2*B*n*w)**2)

def delta(n, w):
  return atan2(2*B*n*w, w_0**2 - n**2*w**2)

def x(nmax, t, w):
  sym_x = a0 + Sum(A(n, w)*cos(n*w*t-delta(n, w)), (n, 1, nmax))
  return sym_x.doit()

```

```python
fig4 = plot(x(1, t, w), (t, 0, 5*T), line_color = 'red', label='n = 1', show=False, adaptive=False)
fig5 = plot(x(5, t, w), (t, 0, 5*T), line_color = 'blue', label='n = 5', show = False, adaptive=False)
fig4.legend = True
fig4.extend(fig5)
fig4.show()
```

```python
fig4 = plot(x(1, t, w2), (t, 0, 5*T), line_color = 'red', label='n = 1', show=False, adaptive=False)
fig5 = plot(x(5, t, w2), (t, 0, 5*T), line_color = 'blue', label='n = 5', show = False, adaptive=False)
fig4.legend = True
fig4.extend(fig5)
fig4.show()
```

```python
fig4 = plot(x(1, t, w3), (t, 0, 5*T), line_color = 'red', label='n = 1', show=False, adaptive=False)
fig5 = plot(x(5, t, w3), (t, 0, 5*T), line_color = 'blue', label='n = 5', show = False, adaptive=False)
fig4.legend = True
fig4.extend(fig5)
fig4.show()
```

```python
fig4 = plot(x(1, t, w4), (t, 0, 5*T), line_color = 'red', label='n = 1', show=False, adaptive=False)
fig5 = plot(x(5, t, w4), (t, 0, 5*T), line_color = 'blue', label='n = 5', show = False, adaptive=False)
fig4.legend = True
fig4.extend(fig5)
fig4.show()
```
