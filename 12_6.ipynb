```python
import whit_phys_util as buc
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
repo = buc.local_repository("https://github.com/cambo2018/Advanced-Dynamics")
```

```python
gamma = .9
w = 2*np.pi
w0 = 1.5*w
beta = w0/4
```

```python
def solver(t, u):
  [phi, phidot] = u
  function = [u[1], gamma*w0**2*cos(w*t) - 2*beta*phidot - w0**2*sin(phi)]
  return function
```

```python
time = [0,6]
tspan = np.linspace(0,6,2837)

y1 = [0, 0]
solution1 = solve_ivp(solver, time, y1, t_eval=tspan)

y2 = [np.pi/2, 0]
solution2 = solve_ivp(solver, time, y2, t_eval=tspan)

y3 = [-np.pi/2, 0]
solution3 = solve_ivp(solver, time, y3, t_eval=tspan)
```

```python
plt.plot(solution1.t, solution1.y[0], label='phi(0) = 0')
plt.plot(solution2.t, solution2.y[0], label='phi(0) = pi/2')
plt.plot(solution3.t, solution3.y[0], label='phi(0) = -pi/2')

plt.xlabel('Time (s)')
plt.ylabel('Angle (rad)')

plt.legend();
```

My results support the claim that all solutions approach the same periodic attractor.
