```python
import whit_phys_util as buc
from sympy import *
from sympy import symbols
```

```python
repo = buc.local_repository("https://github.com/cambo2018/Advanced-Dynamics")
```

```python
t = var('t')
```

```python
x0, omega, v0x, v0y = symbols('x0 omega v0x v0y')
```

```python
omega = 1
x0 = 1
```

```python
x = Function('x')(t)
x = (x0 + v0x*t)*cos(omega*t) + (v0y+omega*x0)*t*sin(omega*t)

y = Function('y')(t)
y = -(x0 + v0x*t)*sin(omega*t)+ t*(v0y+omega*x0)*cos(omega*t)
```

$v_0 = (0,1)$

```python
params = {v0x:0, v0y:1}
plot_parametric((x.subs(params), y.subs(params)), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

$v_0 = (0,0)$

```python
params = {v0x:0, v0y:0}
plot_parametric((x.subs(params), y.subs(params)), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

$v_0 = (0,-1)$

```python
params = {v0x:0, v0y:-1}
plot_parametric((x.subs(params), y.subs(params)), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

$v_0 = (-0.5,-0.5)$

```python
params = {v0x:-.5, v0y:-.5}
plot_parametric((x.subs(params), y.subs(params)), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

$v_0 = (-0.7,-0.7)$

```python
params = {v0x:-.7, v0y:-.7}
plot_parametric((x.subs(params), y.subs(params)), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

$v_0 = (0,-0.1)$

```python
params = {v0x:0, v0y:-.1}
plot_parametric((x.subs(params), y.subs(params)), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```
