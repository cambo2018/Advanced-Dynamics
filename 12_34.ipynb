```python
def sinemap(r,x):
  sinemap = r*np.sin(np.pi*x)
  return sinemap
```

```python
def xt(r,x):
  xt = [x0]
  for i in range(timesteps):
    temp = sinemap(r,xt[i])
    xt.append(temp)
    xstable = xt[400:500]
  return xstable
```

```python
timesteps = 500
rsteps = 10000
r = np.linspace(.6,1,rsteps)

x0 = .1
```

```python
x = []
for i in r:
  x.append(xt(i,x0)) 
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(r, x, '.', markersize=1)
ax.set_xlabel('r')
ax.set_ylabel('x');
```
