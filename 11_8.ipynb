```python
t = linspace(0,30,876)
```

Part 1

```python
#defining constants
v0 = 1
w1 = 1
w2 = w1*sqrt(3)

B1 = 0
B2 = 0
C1 = 0
C2 = -v0/w2
```

```python
#equation of motion
x1 = C2*sin(w2*t)
x2 = C2*sin(w2*t)*(-1)
```

```python
# Plotting solution
fig, ax1 = plt.subplots()

ax1.plot(t, x1, label='x1')
ax1.plot(t, x2, label='x2')

ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Displacent (m)')

ax1.legend()
```

Part 2

```python
#redefining constants
v0 = 1
w1 = 1
w2 = w1*sqrt(3)

B_1 = 0
B_2 = 0
C_1 = v0/2
C_2 = -v0/(2*w2)
```

```python
# new equations of motion
x_1 = C_1*sin(w1*t) + C_2*sin(w2*t)
x_2 = C_1*sin(w1*t) + C_2*sin(w2*t)*(-1)
```

```python
# Plotting solution
fig, ax2 = plt.subplots()

ax2.plot(t, x_1, label='x_1')
ax2.plot(t, x_2, label='x_2')

ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Displacent (m)')

ax2.legend()
```
